### 1. Which establishments have a hygiene score equal to 20?

```python
for doc in establishments.find().limit(5):
    print(doc)
```

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}
results = establishments.find(query)
results = list(results)

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Establishments with a hygiene score of 20: {count}")

# Display the first document in the results using pprint
for result in results:
    pprint(result)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'AddressLine4': {'$regex': 'London'},
    'RatingValue': {'$gte': 4}
        }
results = establishments.find(query)
results = list(results)

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"Establishments in London with a RatingValue greather than or equal to 4: {count}")

# Display the first document in the results using pprint
for result in results:
    pprint(result)
```

```python
# Convert the result to a Pandas DataFrame
high_quality_london_df = pd.DataFrame(results)

# Display the number of rows
num_rows = len(high_quality_london_df)
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows
print(high_quality_london_df.head(10))
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

penang_query = {'BusinessName': 'Penang Flavuors'}

penang_doc = establishments.find_one(penang_query)

#bypass error for lat/long data types 
if penang_doc:
    latitude = float(penang_doc.get('geocode', {}).get('latitude', 0))
    longitude = float(penang_doc.get('geocode', {}).get('longitude', 0))

degree_search = 0.01

query = {
    'RatingValue': 5,
    'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search},
    'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search}
}
sort = [('scores.Hygiene', 1)]
limit = 5

results = list(establishments.find(query).sort(sort).limit(limit))

# Print the results
print("Top 5 establishments with RatingValue of 5, sorted by lowest hygiene score, near 'Penang Flavours':")
for result in results:
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
unhygienic_near_penang_df = pd.DataFrame(results)

# Display the number of rows
num_rows = len(unhygienic_near_penang_df)
print(f"Number of rows in the DataFrame: {num_rows}")

# Display the first 10 rows - confirming only 5 results were stored
print(unhygienic_near_penang_df.head(10))
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': 0 }}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': '$LocalAuthorityName', 'count': { '$sum': 1 }}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': { 'count': -1 }}

# Put the pipeline together
pipeline = [match_query, group_query, sort_values]
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of Local Authorities with hygiene score of 0:", len(results))

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
result_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(result_df))

# Display the first 10 rows of the DataFrame
result_df.head(10)
```
